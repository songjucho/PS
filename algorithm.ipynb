```python
def solution(gems):
    s = set()
    start = 0
    end = len(gems) - 1
    length = len(set(gems))
    
    for i in range(len(gems)):
        for j in range(i, len(gems)):
            s.add(gems[j])
            if len(s) == length and (j - i) < (end - start):
                start = i
                end = j
                break
        s.clear()
    return [start + 1, end + 1]
```

`가중 그래프에서 최소 가중치로 모든 노드 순회`

```python
def solution(n, costs):
    i = 0
    costs.sort(key = lambda x:x[2])
    ans = costs[0][2]
    visited = set([costs[0][0], costs[0][1]])
    costs.pop(0)
    
    while(len(visited) < n):
        #print(visited, costs, ans)
        if costs[i][0] in visited or costs[i][1] in visited:
            if costs[i][0] in visited and costs[i][1] in visited:
                i += 1
                continue
            ans += costs[i][2]
            visited = visited.union(costs[i][:2])
            del costs[i]
            i = 0
            continue
        i += 1

    return ans

solution(6, [[0, 1, 5], [0, 3, 2], [0, 4, 3], [1, 4, 1], [3, 4, 10], [1, 2, 2], [2, 5, 3], [4, 5, 4]])
```

```python
def is2x2(m, n, board):
    answer = []
    for i in range(m-1):
        for j in range(n-1):
            temp = board[i][j]
            if temp == 0:
                continue
            if temp == board[i+1][j] and temp == board[i][j+1] and temp == board[i+1][j+1]:
                answer += [[i, j], [i+1, j], [i, j+1], [i+1, j+1]]
                
    return list(set(map(tuple, answer)))

def removeBlock(m, board, l):
    for i in l:
        board[i[0]].pop(i[1])
    for j in range(m):
        if len(board[j]) < m:
            board[j] += [0] * (m - len(board[j]))
    return board
    
def solution(m, n, board):
    board = list(map(list, zip(*board[::-1])))
    m, n = n, m
    answer = 0
    while(True):
        temp = is2x2(m, n, board)
        if temp == []:
            break
        answer += len(temp)
        board = removeBlock(m, board, sorted(temp, reverse = True))
        
    return answer

solution(3,2, ["AA", "AA", "AB"])
```

```python
import heapq

def dijkstra(graph, start):
    
    distances = {node: float('inf') for node in graph}
    distances[start] = 0
    queue = []
    heapq.heappush(queue, [distances[start], start])
    
    while queue:
        current_distance, current_node = heapq.heappop(queue)
        
        if distances[current_node] < current_distance:
            continue
            
        for adjacent, weight in graph[current_node].items():
            distance = current_distance + weight
            
            if distance < distances[adjacent]:
                distances[adjacent] = distance
                heapq.heappush(queue, [distance, adjacent])
                
    return distances

def solution(n, s, a, b, fares):
    answer = []
    dic = {}
    for i in range(n):
        dic[i+1] = {}

    for j in fares:
        dic[j[0]][j[1]] = j[2]
        dic[j[1]][j[0]] = j[2]
    candidate = dijkstra(dic, s)

    
    answer.append(candidate[a] + candidate[b])
    for i in range(1, n+1):
        if i == s:
            continue
        mid = i
        temp = dijkstra(dic, mid)
        answer.append(candidate[mid] + temp[a] + temp[b])

    return min(answer)

solution(6,4,6,2,[[4, 1, 10], [3, 5, 24], [5, 6, 2], [3, 1, 41], [5, 1, 24], [4, 6, 50], [2, 4, 66], [2, 3, 22], [1, 6, 25]])
```

### binary search - lowerbound
`정렬된 배열에서 target 값이 존재할 때 target 값보다 크거나 같은 값들의 첫번째 위치를 반환한다.`
- arr: 찾고자 하는 리스트
- num: n
- output: n보다 큰 숫자가 나온 위치

### upperbound
`정렬된 배열에서 target 값이 존재할 때 target 값을 초과하는 값들 중 첫번째 위치를 반환한다.`

```python
def upperBoundSearch(arr, target):
    if len(arr) >= 1 and target > arr[-1]:
        return len(arr)
    left, right = 0, len(arr) - 1
    while (left < right):
        mid = (left + right) // 2
        if(arr[mid] <= target):
            left = mid + 1
        else:
            right = mid

    return right


def lowerboundSearch(arr, target):
    if len(arr) == 0 or (len(arr) >= 1 and target > arr[-1]):
        return len(arr)
    left, right = 0, len(arr) - 1
    while (left < right):
        mid = (left + right) // 2
        if(arr[mid] < target):
            left = mid + 1
        else:
            right = mid

    return right

upperBoundSearch([1,2,5,7,8,9], 6)
```

```python
def lowerboundSearch(arr, target):
    if len(arr) == 0 or (len(arr) >= 1 and target > arr[-1]):
        return len(arr)
    left, right = 0, len(arr) - 1
    while (left < right):
        mid = (left + right) // 2
        if(arr[mid] < target):
            left = mid + 1
        else:
            right = mid

    return right

def parseQuery(s):
    string = s.replace("and ", "").split(" ")
    number = int(string[-1])
    return ''.join(string[:-1]), number

def parseInfo(info):
    arr = info.split(' ')
    keys = []
    for i in (arr[0], '-'):
        for j in (arr[1], '-'):
            for k in (arr[2], '-'):
                for l in (arr[3], '-'):
                    keys.append(i+j+k+l)
    return keys, int(arr[-1])

def solution(info, query):
    answer = []
    dic = {}
    
    for l in ('cpp', 'java', 'python', '-'):
        for i in ('backend', 'frontend', '-'):
            for s in ('senior', 'junior', '-'):
                for f in ('pizza', 'chicken', '-'):
                    dic[l+i+s+f] = []

    for l in info:
        keys, score = parseInfo(l)
        for key in keys:
            dic[key].append(score)
    
    for k in dic.keys():
        dic[k].sort()

    for i in query:
        q, num = parseQuery(i)
        scores = dic[q]
        answer.append(len(scores) - lowerboundSearch(scores, num))
    return answer

solution(["java backend junior pizza 150","python frontend senior chicken 210","python frontend senior chicken 150","cpp backend senior pizza 260","java backend junior chicken 80","python backend senior chicken 50"], ["java and backend and junior and pizza 100","python and frontend and senior and chicken 200","cpp and - and senior and pizza 250","- and backend and senior and - 150","- and - and - and chicken 100","- and - and - and - 150"])
```

```python
def longestDecomposition(text):
    strLength = 1
    answer = 0
    while strLength <= len(text):
        if text[:strLength] == text:
            answer += 1
            break
        if text[:strLength] == text[-strLength:]:
            text = text[strLength:-strLength]
            strLength = 0
            answer += 2
        strLength += 1
    return answer
```

### 카카오 공채 광고 삽입

```python
def timeToSec(time):
    h, m, s = map(int, time.split(":"))
    return h * 3600 + m * 60 + s

def secToTime(sec):
    h = str(sec // 3600)
    m = str((sec % 3600) // 60)
    s = str(sec % 60)
    return h.zfill(2)+':'+m.zfill(2)+':'+s.zfill(2)

def solution(play_time, adv_time, logs):
    # step 1
    play_time_sec = timeToSec(play_time)
    adv_time_sec = timeToSec(adv_time)
    logs_start_sec = []
    logs_end_sec = []
    
    for time in logs:
        start, end = time.split("-")
        logs_start_sec.append(timeToSec(start))
        logs_end_sec.append(timeToSec(end))
    
    # step 2
    # total_time[x] = x 시각에 시작된 재생 구간의 개수 – x 시각에 종료된 재생구간의 개수
    total_time = [0 for _ in range(360000)]
    for i in range(len(logs)):
        total_time[logs_start_sec[i]] += 1
        total_time[logs_end_sec[i]] -= 1

    # step 3
    # total_time[x] = 시각 x부터 x+1까지 1초 간의 구간을 포함하는 재생 구간의 개수
    for j in range(1, play_time_sec):
        total_time[j] = total_time[j] + total_time[j - 1]
    
    # step 4
    # total_time[x] = 시각 0부터 x+1까지 x+1초 간의 구간을 포함하는 누적 재생시간
    for k in range(1, play_time_sec):
        total_time[k] = total_time[k] + total_time[k - 1]
    
    # step 5
    answer = float('inf')
    max_time = total_time[adv_time_sec] - 1
    for i in range(adv_time_sec - 1, play_time_sec):
        startTime = total_time[i] - total_time[i - adv_time_sec]
        if startTime > max_time:
            max_time = startTime
            answer = i - adv_time_sec + 1

    return secToTime(answer)

solution("02:03:55","00:14:15",["01:20:15-01:45:14", "00:40:31-01:00:00", "00:25:50-00:48:29", "01:30:59-01:53:29", "01:37:44-02:02:30"])
```

```python
# 파이썬 재귀 제한 푸는 코드
import sys
sys.setrecursionlimit(10**6)
 
class Tree:
    def __init__(self, data, left=None, right=None):
        self.data = data
        self.left = left
        self.right = right
        
preorderList=[]
postorderList=[]
        
def preorder(node,nodeinfo):
    preorderList.append(nodeinfo.index(node.data) + 1)
    if node.left:
        preorder(node.left,nodeinfo)
    if node.right:
        preorder(node.right,nodeinfo)
        
def postorder(node,nodeinfo):
    if node.left:
        postorder(node.left,nodeinfo)
    if node.right:
        postorder(node.right,nodeinfo)
    postorderList.append(nodeinfo.index(node.data)+1)
    
def solution(nodeinfo):
    answer = []
    sortedNodeinfo = sorted(nodeinfo, key=lambda x : (-x[1],x[0]))
    root = None
    
    for node in sortedNodeinfo:
        if not root:
            root = Tree(node)
        else:
            current=root
            while True:
                if node[0] < current.data[0]:
                    if current.left:
                        current = current.left
                        continue
                    else:
                        current.left = Tree(node)
                        break
                if node[0] > current.data[0]:
                    if current.right:
                        current = current.right
                        continue
                    else:
                        current.right = Tree(node)
                        break
                break
    preorder(root,nodeinfo)
    postorder(root,nodeinfo)
    
    return [preorderList] + [postorderList]

solution([[5,3],[11,5],[13,3],[3,5],[6,1],[1,3],[8,6],[7,2],[2,2]])
```

### BOJ 20437

```python
import collections

def stringGame_short(s, count):
    shortest = len(s)
    length = 2
    while length < len(s):
        for i in range(0, len(s) - length + 1):
            if max(collections.Counter(s[i:i+length]).values()) == count:
                shortest = length
                return shortest
        length += 1
    return length

def stringGame_long(s, count):
    length = len(s)
    while length > 0:
        for i in range(0, len(s) - length + 1):
            h = collections.Counter(s[i:i+length])
            if max(h.values()) == count and s[i] == s[i+length-1] and h[s[i]] == count:
                return length
        length -= 1
    return length

n = int(input())
for i in range(n):
    s = input()
    count = int(input())
    if max(collections.Counter(s).values()) < count:
        print(-1)
    else:
        print(stringGame_short(s, count), stringGame_long(s, count))
```

### 프로그래머스 징검다리 건너기

```python
def isOkay(stones, k):
    if ('0,'*k)[:-1] in stones:
        return False
    return True

def check(stones, ans, k):
    arr = ''
    for i in stones:
        if i-ans>=0:
            arr += str(i-ans)+','
        else:
            arr += '0,'
    arr = arr[:-1]
    okay = isOkay(arr, k)
    
    arr2 = ''
    for i in stones:
        if i-ans-1>=0:
            arr2 += str(i-ans-1)+','
        else:
            arr2 += '0,'
    arr2 = arr2[:-1]
    okay2 = isOkay(arr2, k)
    
    return True if okay == True and okay2 == False else False

def solution(stones, k):
    minimum = 1
    maximum = max(stones)
    
    if minimum == maximum or len(stones) == k:
        return maximum
    
    while minimum != maximum - 1:
        ans = (minimum + maximum) // 2
        arr = ''
        for i in stones:
            if i-ans+1>=0:
                arr += str(i-ans+1)+','
            else:
                arr += '0,'
        arr = arr[:-1]

        if isOkay(arr, k):
            minimum = ans
            temp = maximum
        else:
            maximum = ans
            temp = minimum

        if check(stones, minimum, k):
            return minimum+1
        elif check(stones, maximum, k):
            return maximum+1
    return temp
```

### Union-find 최적화 코드
#### Disjoint set 이란?
- Disjoint set은 서로 중복되지 않는 부분 집합들로 나누어진 원소들에 대한 정보를 저장하고 조작하는 자료구조이다.
- 집합을 구현하는 데는 비트벡터, 배열, 연결리스트, 트리 등을 이용할 수 있으나 그 중 가장 효율적인 트리구조를 이용하여 구현한다.

<br>

**make-set(x)** : 초기화 작업, x를 유일한 원소로 하는 새로운 집합을 만든다.

**find(x)** : x가 속한 집합의 대표값(루트 노드)을 반환한다. 즉, x가 어떤 집합에 속해 있는지 찾는 연산이다.

**union(x, y)** : x가 속한 집합과 y가 속한 집합을 합친다.

<br>

`배열이 아닌 해시로 구현하는 방법 고민해보기!`

```python
n = 10
parent = [i for i in range(n)]
rank = [1 for _ in range(n)]

def find(u):
    if u == parent[u]:
        return u
    parent[u] = find(parent[u])
    return parent[u]

def union(u, v):
    u = find(u)
    v = find(v)
    
    if u == v:
        return
    if rank[u] > rank[v]:
        u, v = v, u
    parent[u] = v
    if rank[u] == rank[v]:
        rank[v] += 1
union(1, 4)
union(1, 9)
find(9)
```

```python
from collections import deque
import itertools

n, m = map(int, input().rstrip().split())
dic = {}

def dfs_paths(graph, start, goal):
    stack = [(start, [start])]
    result = []

    while stack:
        n, path = stack.pop()
        if n == goal:
            result.append(path)
        else:
            for m in set(graph[n]) - set(path):
                stack.append((m, path + [m]))
    return result

for i in range(n):
    dic[i] = []

for i in range(m):
    a, b = map(int, input().rstrip().split())
    dic[a].append(b)
    
comb = list(itertools.combinations([i for i in range(n)], 2))
ans = 0
for i in comb:
    path = dfs_paths(dic, i[0], i[1])
    if len(path) > 0 and len(path[0]) >= 5:
        print(1)
        break
else:
    print(0)
```

```python
from collections import deque

def solution(board, moves):
    answer = 0
    new_board = deque()
    stack = []
    for i in range(len(board)):
        new_board.append(deque([j[i] for j in board if j[i] != 0]))
    for i in moves:
        if len(new_board[i-1]) == 0:
            continue
        popped = new_board[i-1].popleft()
        if len(stack) > 0 and popped == stack[-1]:
            answer += 2
            stack.pop()
        else:
            stack.append(popped)  
    return answer
```

### 프로그래머스 다단계 칫솔 판매 풀이

```python
def solution(enroll, referral, seller, amount):
    result = {}
    tree = {}
    for i in range(len(enroll)):
        result.setdefault(enroll[i], 0)
        tree.setdefault(enroll[i], {})
        if referral[i] in tree.keys():
            tree[enroll[i]] = {referral[i]: tree[referral[i]]}
        else:
            tree[enroll[i]] = {}
    for i in range(len(seller)):
        name = seller[i]
        money = amount[i] * 100
        result[name] += money
        key = name
        while True:
            divided = money // 10
            result[key] -= divided
            if tree[key] == {} or divided < 1:
                break
            key = list(tree[key].keys())[0]
            result[key] += divided
            money = divided
    return list(result.values())
```

```python
import itertools

def getColumn(arr, indexes):
    ans = []
    for i in arr:
        tmp = []
        for j in range(len(i)):
            if j in indexes:
                tmp.append(i[j])
        ans.append(tuple(tmp))
    return ans

def removeCombIndex(indexes, idx):
    new = []
    if len(indexes) < 2:
        return new
    for i in indexes[1:]:
        if len(set(indexes[idx]) - set(i)) != 0:
            new.append(i)
    return new

def solution(relation):
    answer = 0
    n = len(relation[0])
    arr = [i for i in range(len(relation[0]))]
    comb = []
    for i in range(1, n+1):
        comb += list(itertools.combinations(arr, i))
    idx = 0
    comb = sorted(comb, key=lambda x:len(x))
    while len(comb) > 0 and idx < len(comb):
        arr = getColumn(relation, comb[idx])
        #print(arr, set(arr))
        if len(arr) == len(set(arr)):
            answer += 1
            comb = removeCombIndex(comb, idx)
        else:
            idx += 1
    return answer
        

solution([['a',1,4], [2, 1,5], ['a',2, 4]])
```

```python
def solution(ages, wires):
    off_list = []
    N = len(ages)

    # 0. 최대 전선 길이 리스트 만들기
    max_length = [0] * N
    for wire in wires:
        if wire[2] > max_length[wire[1] - 1]:
            max_length[wire[1] - 1] = wire[2]

    while len(off_list) < N:
        # 1. 전체 수명 -= 1
        for i in range(N):
            # 1-1. 이미 중단되어 있는 발전기는 패스
            if ages[i] < 0:
                continue

            # 1-2. 작동하는 발전기라면 수명 -= 1
            elif ages[i] > 0:
                ages[i] -= 1

            # 이제 막 멈춘 발전기
            if ages[i] == 0:
                off_list.append(i + 1)
                ages[i] = -1
                max_length[i] = -2

        # 2. 중단된 발전기가 공급하는 전력(length) -= 1
        for off in off_list:
            for i in range(len(wires)):
                # 2-1. start가 중단된 발전기라면
                if off == wires[i][0] and wires[i][2] > -1:
                    wires[i][2] -= 1
                    max_length[wires[i][1] - 1] -= 1

        # 3. 최대 length가 0이 된 발전기 중단
        for i in range(N):
            if max_length[i] == -1:
                off_list.append(i + 1)
                max_length[i] = -2
                ages[i] = -1

    return off_list

solution([35,25,3,8,7], [[1,2,5],[2,1,5],[1,3,2],[3,4,2],[3,5,20],[4,5,1]])
```

### LCA 예제

```python
import sys
sys.setrecursionlimit(int(1e5)) # 런타임 오류를 피하기 위한 재귀 깊이 제한 설정

n = int(input())

parent = [0] * (n + 1) # 부모 노드 정보
d = [0] * (n + 1) # 각 노드까지의 깊이
c = [0] * (n + 1) # 각 노드의 깊이가 계산되었는지 여부
graph = [[] for _ in range(n + 1)] # 그래프(graph) 정보

for _ in range(n - 1):
    a, b = map(int, input().split())
    graph[a].append(b)
    graph[b].append(a)

# 루트 노드부터 시작하여 깊이(depth)를 구하는 함수
def dfs(x, depth):
    c[x] = True
    d[x] = depth
    for y in graph[x]:
        if c[y]: # 이미 깊이를 구했다면 넘기기
            continue
        parent[y] = x
        dfs(y, depth + 1)

# A와 B의 최소 공통 조상을 찾는 함수
def lca(a, b):
    # 먼저 깊이(depth)가 동일하도록
    while d[a] != d[b]:
        if d[a] > d[b]:
            a = parent[a]
        else:
            b = parent[b]
    # 노드가 같아지도록
    while a != b:
        a = parent[a]
        b = parent[b]
    return a

dfs(1, 0) # 루트 노드는 1번 노드

m = int(input())

for i in range(m):
    a, b = map(int, input().split())
    print(lca(a, b))
```

```python
def createTable(string, stlen):
    table = [-1] + [0 for _ in range(stlen)]
    j = 0
    for k in range(1, stlen):
        while string[j] != string[k] and j > 0:
            j = table[j]
        if string[j] == string[k]:
            j += 1
            table[k+1] = j
    return table

def kmp(string, pattern):
    stlen = len(string)
    patlen = len(pattern)
    distance, idx, cnt = 0, 0, 0
    
    table = createTable(pattern, patlen)
    while True:
        idx = 0
        if idx + distance + patlen > stlen:
            break
        while string[idx + distance] == pattern[cnt]:
            cnt += 1
            idx += 1
            if cnt == patlen:
                return True
        distance += cnt - table[cnt]
        cnt = 0
    return False

s = input()
p = input()

if kmp(s, p) == True:
    print(1)
else:
    print(0)
```

```python
import sys

def get_time(n:int,m:int,t_list:list) -> int:
    # 이분 탐색
    left, right = 0, n * max(t_list)
    t = None
    while left <= right:
        mid = (left + right) // 2
        cnt = m # 초기 m 명 태우고 시작
        # mid 분 동안 몇 명 태울 수 있는지 계산하기 
        for i in range(m):
            cnt += mid // t_list[i]
        if cnt >= n: # n명을 태울 수 있는 시간 기록하고 더 짧은 시간 없는지 찾아보기
            t = mid
            right = mid - 1
        else: # n명을 태우기 위해서는 시간이 더 필요
            left = mid + 1
    return t


if __name__ == "__main__":
    n, m = map(int, sys.stdin.readline().split())
    t_list = list(map(int, sys.stdin.readline().split()))

    if n < m:
        print(n)
    else:
        t = get_time(n, m, t_list)
        # t - 1에 탑승한 아이들의 숫자를 계산합니다.
        cnt = m
        for i in range(m):
            cnt += (t - 1) // t_list[i]
        # t분에 탑승할 수 있는 놀이기구를 앞에서부터 확인한다.
        for i in range(m):
            if t % t_list[i] == 0:  # 탈 수 있는 놀이기구 발견
                cnt += 1
            # 탑승한 아이가 n 명이 되었을 때 종료합니다. 
            if cnt == n:
                print(i + 1)
                break
```

### 프로그래머스 퍼즐 조각 채우기

```python
from collections import deque

def bfs(board, flag):
    n = len(board)
    dx = [0, 1, 0, -1]
    dy = [1, 0, -1, 0]
    visited = [[0] * n for _ in range(n)]
    q = deque()
    blocks = []
    
    for i in range(n):
        for j in range(n):
            if visited[i][j] == 1 or board[i][j] != flag:
                continue
            q.append([i,j])
            visited[i][j] = 1
            minx, miny = n, n
            maxx, maxy = 0, 0
            tmp = []
            while q:
                x, y = q.popleft()
                tmp.append([x, y])
                minx = min(minx, x)
                miny = min(miny, y)
                maxx = max(maxx, x)
                maxy = max(maxy, y)
                for k in range(4):
                    xx = x + dx[k]
                    yy = y + dy[k]
                    if 0<=xx<n and 0<=yy<n and visited[xx][yy] == 0 and board[xx][yy] == flag:
                        q.append([xx, yy])
                        visited[xx][yy] = 1
            # 구간만 따로 저장
            blocks.append([row[miny:maxy+1] for row in board[minx:maxx+1]])
    return blocks   

def rotated(a):
    n = len(a)
    m = len(a[0])
    result = [[0]* n for _ in range(m)]
    for i in range(n):
        for j in range(m):
            result[j][n-i-1] = a[i][j]
    return result
    
def solution(game_board, table):
    ans = 0
    n = len(table)
    for i in range(n):
        for j in range(n):
            if table[i][j] == 1:
                table[i][j] = 0
            else:
                table[i][j] = 1
    
    boardBlocks = bfs(game_board, 0)
    tableBlocks = bfs(table, 0)
    
    for i in tableBlocks:
        num = sum(i, []).count(0)
        flag = False
        r = i
        for _ in range(4):
            r = rotated(r)
            for j in boardBlocks:
                if r == j:
                    boardBlocks.remove(j)
                    ans += num
                    flag = True
                    break
            if flag == True:
                break
    return ans
```

### 프로그래머스  직업군 추천하기

```python
from collections import defaultdict

def solution(table, languages, preference):
    scoreTable = {}
    answer = []
    maxScore = 0
    for i in table:
        splitted = i.split()
        scoreTable[splitted[0]] = {}
        score = 5
        for j in splitted[1:]:
            scoreTable[splitted[0]][j] = score
            score -= 1

    t = {}
    for i in range(len(languages)):
        t[languages[i]] = preference[i]
    
    for i in scoreTable.keys():
        total = 0
        for j in languages:
            scoreTable[i].setdefault(j, 0)
            total += scoreTable[i][j] * t[j]
        if total > maxScore:
            answer = [i]
            maxScore = total
        elif total == maxScore:
            answer.append(i)

    return sorted(answer)[0]
```

```python
import datetime

def solution(n, t, m, timetable):
    times = []
    for i in timetable:
        if i == "23:59":
            continue
        times.append(datetime.datetime.strptime(i, '%H:%M'))
    times.sort()
    busTime = []
    startTime = datetime.datetime.strptime('09:00', '%H:%M')
    for i in range(n):
        time = startTime + datetime.timedelta(seconds=i*t*60)
        busTime.append([time, 0]) #시간과 현재 인원수
    
    while times and busTime != []:
        #print(times, busTime)
        if times[0] <= busTime[0][0]:
            busTime[0][1] += 1
            if busTime[0][1] == m:
                if len(busTime) == 1:
                    return (times[0]-datetime.timedelta(seconds=60)).strftime("%H:%M")
                busTime.pop(0)
            times.pop(0)

        else:
            if len(busTime) == 1:
                return busTime[-1][0].strftime("%H:%M")
            busTime.pop(0)

    return busTime[-1][0].strftime("%H:%M")

solution(1,1,1,["18:00"]*55)
```

```python
from collections import deque

def solution(board):
    N = len(board)
    dx = [1, 0, -1, 0]
    dy = [0, 1, 0, -1]
    
    q = deque()
    q.append([[0, 0], [0, 1], 0, 0]) #시작좌표, 방향(가로:0, 세로:1), 시간
    visited = [[-1] * N for _ in range(N)]
    visited[0][0] = 0
    
    while q:
        first, second, direction, time = q.popleft()
        if second == [N-1, N-1]:
            return time
        for i in range(4):
            fx = first[0] + dx[i]
            fy = first[1] + dy[i]
            sx = second[0] + dx[i]
            sy = second[1] + dy[i]
            if 0<=fx<N and 0<=fy<N and 0<=sx<N and 0<=sy<N and board[fx][fy] != 1 and board[sx][sy] != 1 and visited[fx][fy] != direction and visited[fx][fy] != 2:
                q.append([[fx, fy], [sx, sy], direction, time+1])
                if visited[fx][fy] == -1:
                    visited[fx][fy] = direction
                else:
                    visited[fx][fy] = 2
        x1, y1 = first
        x2, y2 = second
        if direction == 0 and x1 >= 1 and board[x1-1][y1] != 1 and board[x2-1][y2] != 1:
            if visited[x2-1][y2] != 2 and visited[x2-1][y2] != 1:
                q.append([[x2-1, y2], [x2, y2], 1, time+1])
                if visited[x2-1][y2] == -1:
                    visited[x2-1][y2] = 1
                else:
                    visited[x2-1][y2] = 2
            if visited[x1-1][y1] != 2 and visited[x1-1][y1] != 1:
                q.append([[x1-1, y1], [x1, y1], 1, time+1])
                if visited[x1-1][y1] == -1:
                    visited[x1-1][y1] = 1
                else:
                    visited[x1-1][y1] = 2
        if direction == 0 and x1<N-1 and board[x1+1][y1] != 1 and board[x2+1][y2] != 1:
            if visited[x2+1][y2] != 2 and visited[x2+1][y2] != 1:
                q.append([[x2, y2], [x2+1, y2], 1, time+1])
                if visited[x2][y2] == -1:
                    visited[x2][y2] = 1
                else:
                    visited[x2][y2] = 2
            if visited[x1+1][y1] != 2 and visited[x1+1][y1] != 1:
                q.append([[x1, y1], [x1+1, y1], 1, time+1])
                if visited[x1][y1] == -1:
                    visited[x1][y1] = 1
                else:
                    visited[x1][y1] = 2
        if direction == 1 and y1 >= 1 and board[x1][y1-1] != 1 and board[x2][y2-1] != 1:
            if visited[x1][y1-1] != 2 and visited[x1][y1-1] != 0:
                q.append([[x1, y1-1], [x1, y1], 0, time+1])
                if visited[x1][y1-1] == -1:
                    visited[x1][y1-1] = 0
                else:
                    visited[x1][y1-1] = 2
            if visited[x2][y2-1] != 2 and visited[x2][y2-1] != 0:
                q.append([[x2, y2-1], [x2, y2], 0, time+1])
                if visited[x2][y2-1] == -1:
                    visited[x2][y2-1] = 0
                else:
                    visited[x2][y2-1] = 2
        if direction == 1 and y1<N-1 and board[x1][y1+1] != 1 and board[x2][y2+1] != 1:
            if visited[x1][y1] != 2 and visited[x1][y1] != 0:
                q.append([[x1, y1], [x1, y1+1], 0, time+1])
                if visited[x1][y1] == -1:
                    visited[x1][y1] = 0
                else:
                    visited[x1][y1] = 2
            if visited[x2][y2] != 2 and visited[x2][y2] != 0:
                q.append([[x2, y2], [x2, y2+1], 0, time+1])
                if visited[x2][y2] == -1:
                    visited[x2][y2] = 0
                else:
                    visited[x2][y2] = 2
    return None

solution([[0, 0, 0, 1, 1],[0, 0, 0, 1, 0],[0, 1, 0, 1, 1],[1, 1, 0, 0, 1],[0, 0, 0, 0, 0]])
```

### 프로그래머스 표 편집

```python
import copy

# Node 클래스 정의
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None


# LinkedList 클래스 (자료구조) 정의
class LinkedList:

    # 초기화 메소드
    def __init__(self):
        self.head = Node(None)
        self.current = Node(None)
        self.tail = Node(None)
        
        self.head.prev = None
        self.head.next = self.tail
        self.tail.prev = self.head
        self.tail.next = None


        self.num_of_data = 0

    # append 메소드 (insert - 맨 뒤에 노드 추가, tail과 node의 next, 데이터 개수 변경)
    def append(self, data):
        new_node = Node(data)
        new_node.prev = self.current
        new_node.next = None
        self.current.next = new_node
        
        self.before = self.current
        self.tail.prev = self.current
        self.current = new_node
        self.tail.next = new_node
        self.tail = new_node

        self.num_of_data += 1

    # delete 메소드 (delete - current 노드 삭제, 인접 노드의 current, next 변경, 데이터 개수 변경)
    def delete(self):
        pop_data = self.current.data

        if self.current is self.tail:
            self.tail = self.before
        
        self.before.next = self.current.next
        if self.current.next != None:
            self.current.next.prev = self.before
        self.current = self.before # 중요 : current가 next가 아닌 before로 변경된다.

        self.num_of_data -= 1

        return pop_data

    # first 메소드 (search1 - 맨 앞의 노드 검색, before, current 변경)
    def first(self):
        if self.num_of_data == 0: # 데이터가 없는 경우 첫번째 노드도 없기 때문에 None 리턴
            return None

        self.before = self.head
        self.current = self.head.next

        return self.current.data

    # next 메소드 (search2 - current 노드의 다음 노드 검색, 이전에 first 메소드가 한번은 실행되어야 함)
    def next(self):
        if self.current.next == None:
            return None

        self.before = self.current
        self.current = self.current.next

        return self.current.data
    
    def previous(self):
        if self.current == self.head:
            return None
        self.current = self.before
        print("데이터", self.before.data)
        print("이전", self.before.prev.data)
        self.before = self.before.prev
        print("이후", self.before.prev.data)
        
        return self.current.data
        

    def size(self):
        return self.num_of_data

def solution(n, k, cmd):
    linkedList = LinkedList()
    stack = []
    answer = ''
    for i in range(1, n+1):
        linkedList.append(i)
    linkedList.first()
    for _ in range(k):
        linkedList.next()
    
    for i in cmd:
        if i.startswith("U"):
            u, x = i.split()
            for _ in range(int(x)):
                linkedList.previous()
        elif i.startswith("D"):
            d, x = i.split()
            for _ in range(int(x)):
                linkedList.next()
        elif i == "C":
            stack.append(copy.deepcopy(linkedList))
            linkedList.delete()
        else:
            linkedList = stack.pop()
        print(linkedList.current.data)
    linkedList.first()
    for _ in range(linkedList.num_of_data):
        print(linkedList.next())
    return answer

solution(8, 2, ["D 2","C","U 3","C","D 4","C","U 2","Z","Z"])
```
